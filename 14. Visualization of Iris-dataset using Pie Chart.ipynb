```python
import pandas as pd 
from matplotlib import pyplot as plt 
 
t = pd.read_csv("IRIS.csv") 
sv=t.groupby("species")["petal_length"].mean() 
 
plt.pie(sv,labels=sv.index,startangle=90,shadow=True,autopct="%1.0f%%") 
plt.title("Petal length of Species Average Visualization") 
plt.show ()
```

```python
import pandas as p
import matplotlib.pyplot as m
d=p.read_csv("IRIS.csv")
s=d.groupby('species')['sepal_length'].mean()
m.pie(s,labels=s.index,startangle=70,shadow=True,autopct='1%.f%%')
m.show()
```
