```python
import nltk
import matplotlib.pyplot as m
import pandas as p
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

f=open('akshay.txt')
text=f.read()
w=word_tokenize(text)
print(' before removal stop words word count',len(w))
stop_w=nltk.corpus.stopwords.words('english')

remove_stopw=[]
for i in w:
    if i not in stop_w:
        remove_stopw.append(i)
print('\n after removed stop word \n',len(remove_stopw))
freq=FreqDist(remove_stopw)
freqs=p.Series(dict(freq))

m.figure(figsize=(18,10))
freqs.plot(kind='bar')
m.show()
```
