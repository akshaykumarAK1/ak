```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
df =pd.read_csv('IRIS.csv')

# Create the pairplot with a specified figure size
sns.pairplot(df, hue="species", height=3)  # Adjust height as needed
plt.suptitle('Multivariate Plot', y=1.02)

# Display the plot
plt.show()
```
